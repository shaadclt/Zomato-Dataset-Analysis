# Data Cleaning

## Dropping Redundant columns

```python
data.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1,inplace=True)
data.head()
```

```python
data.info()
```

## Dropping Duplicates

```python
data.drop_duplicates(inplace=True)
data.shape
```

## Cleaning Rate Column

```python
data['rate'].unique
```

### Removing 'NEW','-' and '/5' from rate column

```python
def handle_rate(value):
    if value=='NEW' or value=='-':
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
data['rate'] = data['rate'].apply(handle_rate)
data['rate'].head()
```

### Filling Null Values in Rate Column with Mean

```python
data['rate'].isnull().sum()
```

```python
data['rate'].fillna(data['rate'].mean(),inplace=True)
data['rate'].isnull().sum()
```

```python
data.info()
```

## Dropping Null Values from all columns

```python
data.dropna(inplace=True)
data.head()
```

## Renaming columns for usability

```python
data.rename(columns = {'approx_cost(for two people)': 'cost2plates','listed_in(type)':'type'},inplace=True)
data.head()
```
