# Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

# Loading dataset

```python
data = pd.read_csv(r'data/zomato.csv')
```

# Data Cleaning

## Dropping Redundant columns

```python
data.drop(['url','address','phone','menu_item','dish_liked','reviews_list','listed_in(city)'],axis=1,inplace=True)
data.head()
```

```python
data.info()
```

## Dropping Duplicates

```python
data.drop_duplicates(inplace=True)
data.shape
```

## Cleaning Rate Column

```python
data['rate'].unique
```

### Removing 'NEW','-' and '/5' from rate column

```python
def handle_rate(value):
    if value=='NEW' or value=='-':
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
data['rate'] = data['rate'].apply(handle_rate)
data['rate'].head()
```

### Filling Null Values in Rate Column with Mean

```python
data['rate'].isnull().sum()
```

```python
data['rate'].fillna(data['rate'].mean(),inplace=True)
data['rate'].isnull().sum()
```

```python
data.info()
```

## Dropping Null Values from all columns

```python
data.dropna(inplace=True)
data.head()
```

## Renaming columns for usability

```python
data.rename(columns = {'approx_cost(for two people)': 'cost2plates','listed_in(type)':'type'},inplace=True)
data.head()
```

## Cleaning 'cost2plates' Column

```python
data['cost2plates'].unique
```

### Removing ',' from 'cost2plate' column

```python
def handle_comma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

data['cost2plates'] = data['cost2plates'].apply(handle_comma)
data['cost2plates'].head()
```

```python
data['cost2plates'].tail()
```

## Cleaning 'rest_type' Column

```python
rest_types = data['rest_type'].value_counts()
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

### Making rest_types less 1000 frequency as others

```python
def handle_rest_type(value):
    if value in rest_types_lessthan1000:
        return 'others'
    else:
        return value
data['rest_type'] = data['rest_type'].apply(handle_rest_type)
data['rest_type'].value_counts()
```

## Cleaning 'location' column

```python
locations = data['location'].value_counts()
locations
```

```python
locations_lessthan300 = locations[locations<300]
locations_lessthan300
```

## Making locations less 300 frequency as others

```python
def handle_location(value):
    if value in locations_lessthan300:
        return 'others'
    else:
        return value
data['location'] = data['location'].apply(handle_location)
data['location'].value_counts()
```

## Cleaning 'cuisines' column

```python
cuisines = data['cuisines'].value_counts()
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
def handle_cuisines(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value
data['cuisines'] = data['cuisines'].apply(handle_cuisines)
data['cuisines'].value_counts()
```

## Grouping Type of Restaurant location wise

```python
data3 = data.groupby(['location','type'])['name'].count()
data3.to_csv('location_type.csv')
data3 = pd.read_csv('location_type.csv')
data3 = pd.pivot_table(data3,values=None,index=['location'],columns=['type'],fill_value=0,aggfunc=np.sum).reset_index()
data3.columns = ['location', 'Buffet', 'Cafes', 'Delivery', 'Desserts', 'Dine-out', 'Drinks & nightlife', 'Pubs and bars']
data3['Total'] = data3[['Buffet', 'Cafes', 'Delivery', 'Desserts', 'Dine-out', 'Drinks & nightlife', 'Pubs and bars']].sum(axis=1)
data3
```

```python
data3_perc = data3.select_dtypes('number').div(data3['Total']*0.01, axis=0)
data3.update(data3_perc)
data3.drop('Total', axis=1, inplace=True)
data3.set_index('location', inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(12,7))
bottom = np.zeros(len(data3))

# Create each bar with values using a loop
for i, col in enumerate(data3.columns):
    ax.barh(data3.index, data3[col], left=bottom, label=col)
    bottom+=data3[col].values

# Calculate total values for positioning the total value of stacked bar (100)
totals = data3.sum(axis=1)
x_offset = 4

# Assigning total length of the bar as label (annotation)
for i, t in enumerate(totals):
    ax.text(t + x_offset, i, round(t), va='center', weight='bold')

# Applying labels (annotation) for the middle bars
for bar in ax.patches:
    width = bar.get_width()
    if width > 3:  # Only label visible bars
        ax.text(
            bar.get_x() + width / 2,
            bar.get_y() + bar.get_height() / 2,
            round(width),
            ha='center',
            va='center',
            color='white',
            weight='bold',
            size=8
        )

# Chart control commands
plt.xlim(0,130)
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
